```python
import random

with open("szavak.txt", "r", encoding="utf-8") as file:
    words = file.read().splitlines()


def kivlasztott_szo(word_list, word_length):
    words = [w for w in word_list if len(w) == word_length]
    if len(words) == 0:
        return random.choice(word_list)
    return random.choice(words)


def tippelj(elerheto_betuk, titkos, elet):
    találat = input("\ntippelt szám: ")
    while not találat.isalpha():
        találat = input("\ntippelj egy betűt \n\ntippelj mégegyszer: ")
    while len(találat) > 1:
        találat = input("\nEgy időben csak egy betűt tippelhetsz \n\tippelj mégegyszer: ")
    while találat not in elerheto_betuk:
        találat = input("\nEzt a betűt már egyszer kitaláltad \n\ntippelj mégegyszer: ")
    if találat not in titkos.values():
        elet = elet - 1
        if elet > 1:
            print("\nHibás találat! Már csak: " + str(elet) + " élet maradt!")
            return találat, elet
        else:
            print("\nHibás találat! Már csak " + str(elet) + " élet maradt!")
            return találat, elet
    else:
        return találat, elet


def get_status(status, titkos, jelenlegi_találat):
    for k, v in zip(titkos.keys(), titkos.values()):
        if jelenlegi_találat == v:
            status[k] = v
            print("\nMegfelelő találat!\n")
    print(''.join(status))
    return status


def track_találates(jelenlegi_találat, elerheto_betuk):
    elerheto_betuk = elerheto_betuk.replace(jelenlegi_találat, '')
    return elerheto_betuk


def read_number(message):
    while True:
        again = int(input(message))
        while not isinstance(again, int):
            again = int(input(message))
        return again


def szo_hossza():
    while True:
        again = input("\nSzó hossza: random(r) vagy választható(v)\n")
        while again.lower() not in ["r", "v"]:
            again = input("\nSzó hossza: random(r) vagy választható(v)\n")
        if again.lower() == "r":
            return 0
        else:
            return read_number("\nMilyen hosszú legyen a szó?\n")


def game():
    szohossz = szo_hossza()
    elet = 5
    eletek_szama = read_number("\nMennyi élet legyen?\n")
    if eletek_szama > 0:
      elet = eletek_szama
    titkos = kivlasztott_szo(words, szohossz)
    mydict = {}
    for x in range(len(titkos)):
        mydict.update({x: titkos[x]})
    titkos = mydict
    print(f'\Játsz akasztófát!\n\nA te kiválasztott szavad {str(len(titkos))} betű hosszú.')
    status = ['_'] * len(titkos)
    print(*status)
    elerheto_betuk = "aábcdeéfghiíjklmnoóöőpqrstuúüűvwxyz"
    while list(status) != list(titkos.values()):
        jelenlegi_találat, elet = tippelj(elerheto_betuk, titkos, elet)

        status = get_status(status, titkos, jelenlegi_találat)
        elerheto_betuk = track_találates(jelenlegi_találat, elerheto_betuk)
        if status == list(titkos.values()):
            print("\ntalálat \n\nNyertel!")
            break
        if elet == 0:
            print("\nKifogytál a probálkozási lehetőségekből, Vesztettél")
            break
        continue


if __name__ == "__main__":
    game()
    while True:
        again = input("\nAkarsz újra játszni Igen(I) vagy Nem(N)\n")
        while again.lower() not in ["y", "yes", "n", "no"]:
            again = input("\nKérlek írd be hogy Igen(I) vagy Nem(N): \n")
        if again.lower() == "yes" or again.lower() == "y":
            game()
        else:
            print("\nKöszönjük a játékot!")
            break
```
