```python
#Aoc day4/1 Ã©s day4/2

def get_row(line):
    row = []
    line_parts = line.split(' ')
    for part in line_parts:
        if part != '':
            row.append(int(part))
    return row

def mark(matrix, number):
    for i in range(0, len(matrix)):
        for j in range(0, len(matrix[0])):
            if matrix[i][j] == number:
                matrix[i][j] = -1

def won(matrix):
    for i in range(0, len(matrix)):
        all_marked = True
        for j in range(0, len(matrix[0])):
            if matrix[i][j] != -1:
                all_marked = False
        if all_marked:
            return True
        
    for i in range(0, len(matrix[0])):
        all_marked = True
        for j in range(0, len(matrix)):
            if matrix[j][i] != -1:
                all_marked = False
        if all_marked:
            return True
        
    return False

def sum_unmarked(matrix):
    matrix_sum = 0
    for i in range(0, len(matrix)):
        for j in range(0, len(matrix[0])):
            if matrix[i][j] != -1:
                matrix_sum = matrix_sum + matrix[i][j]
    return matrix_sum

    
lines = []

with open('input4.txt', 'r') as f:
    for line in f:
        line = line.rstrip()
        if line != '':
            lines.append(line)

numbers = []
boards = []

first_row_strings = lines[0].split(',')
for string in first_row_strings:
    numbers.append(int(string))

for i in range(1, len(lines), 5):
    matrix = [
        get_row(lines[i]),
        get_row(lines[i+1]),
        get_row(lines[i+2]),
        get_row(lines[i+3]),
        get_row(lines[i+4])
    ]
    boards.append(matrix)
    
for number in numbers:
    for matrix in boards:
        if matrix[0][0] == -2:
            continue
        mark(matrix, number)
        matrix_won = won(matrix)
        if matrix_won:
            print(sum_unmarked(matrix)*number)
            matrix[0][0] = -2
        
#for matrix in boards:
    #print(matrix)
```
