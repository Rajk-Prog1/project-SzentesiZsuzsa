```python
import random

with open("szavak.txt", "r") as file:
    words = file.read().splitlines()


def kivlasztott_szo(word_list):
    return random.choice(word_list)


def tippelj(elerheto_betuk, titkos, elet):
    találat = input("\ntippelt szam: ")
    while not találat.isalpha():
        találat = input("\ntippelj egy betűt \n\ntippelj megegyszer: ")
    while len(találat) > 1:
        találat = input("\nEgy idoben csak egy betut tippelhetsz \n\tippelj megegyszer: ")
    while találat not in elerheto_betuk:
        találat = input("\nEzt a betut mar egyszer kitalaltad \n\ntippelj megegyszer: ")
    if találat not in titkos.values():
        elet = elet - 1
        if elet > 1:
            print("\nHibas találat! Mar csak: " + str(elet) + " elet maradt!")
            return találat, elet
        else:
            print("\nHibas találat! Mar csak " + str(elet) + " elet maradt!")
            return találat, elet
    else:
        return találat, elet


def get_status(status, titkos, jelenlegi_találat):
    for k, v in zip(titkos.keys(), titkos.values()):
        if jelenlegi_találat == v:
            status[k] = v
            print("\nMegfelelo találat!\n")
    print(''.join(status))
    return status


def track_találates(jelenlegi_találat, elerheto_betuk):
    elerheto_betuk = elerheto_betuk.replace(jelenlegi_találat, '')
    return elerheto_betuk


def game():
    titkos = kivlasztott_szo(words)
    mydict = {}
    for x in range(len(titkos)):
        mydict.update({x: titkos[x]})
    titkos = mydict
    print(f'\Jatsz akasztofat!\n\nA te kivalasztott szavad{str(len(titkos))} letters long.')
    status = ['_'] * len(titkos)
    print(*status)
    elerheto_betuk = "abcdefghijklmnopqrstuvwxyz"
    elet = 5
    while list(status) != list(titkos.values()):
        jelenlegi_találat, elet = tippelj(elerheto_betuk, titkos, elet)

        status = get_status(status, titkos, jelenlegi_találat)
        elerheto_betuk = track_találates(jelenlegi_találat, elerheto_betuk)
        if status == list(titkos.values()):
            print("\ntalálat \n\nNyertel!")
            break
        if elet == 0:
            print("\nKifogytal a probalkozasi lehetosegekbol, Vesztettel")
            break
        continue


if __name__ == "__main__":
    game()
    while True:
        again = input("\nAkarsz ujra jatszni Igen(I) vagy Nem(N)\n")
        while again.lower() not in ["y", "yes", "n", "no"]:
            input("\nPlease input Yes(Y) or No(N): \n")
        if again.lower() == "yes" or again.lower() == "y":
            game()
        else:
            print("\nThanks for playing!")
            break

```
